```python
import numpy as np
import matplotlib.pyplot as plt
import os
import fnmatch
from pathlib import Path
```

```python
def read_csv_matrix(csv_file):
    return np.loadtxt(csv_file, delimiter=';')
```

## Task 1

```python
data = Path('data')
files = os.listdir('data')
task1_data = sorted(fnmatch.filter(files,'task1*'))
# group for participants
participants = []
for name in task1_data:
    p = '_'.join(name.split('_')[1:3])
    if p not in participants:
        participants.append(p)
```

```python
fig, axs = plt.subplots(len(participants), 4, figsize=(12,3*len(participants)))
for i,p in enumerate(participants):
    j = 0
    for f in (500,1000):
        for reim in ('real','imag'):
            h = read_csv_matrix(data / f'task1_{p}_CSM_{reim}_f_{f}.csv')
            im = axs[i,j].imshow(h)
            fig.colorbar(im, ax=axs[i,j])
            axs[i,j].set_title( f'CSM {reim} {f} Hz\n{p}')
            j += 1            
plt.tight_layout()
plt.show()
```

## Task 2

```python
fig, axs = plt.subplots(len(participants), 4, figsize=(12,3*len(participants)))
for i,p in enumerate(participants):
    j = 0
    for f in (500,1000,2000,4000):
            h = 10*np.log10(read_csv_matrix(data / f'task2_{p}_PSF_f_{f}.csv'))
            im = axs[i,j].imshow(h.T, vmin=-30, extent=(-1,1,-1,1))
            fig.colorbar(im, ax=axs[i,j])
            axs[i,j].set_title( f'PSF {f} Hz\n{p}')
            j += 1            
plt.tight_layout()
plt.show()
```

## Task 3

```python
fig, axs = plt.subplots(len(participants), 4, figsize=(12,3*len(participants)))
for i,p in enumerate(participants):
    j = 0
    for f in (500,1000,2000,4000):
            h = read_csv_matrix(data / f'task3_{p}_CBF_f_{f}.csv')
            hmax = np.nanmax(h)
            im = axs[i,j].imshow(h.T, extent=(-2,2,-1,1), vmin=hmax-30) 
            fig.colorbar(im, ax=axs[i,j], orientation='horizontal')
            axs[i,j].set_title( f'CBF {f} Hz\n{p}')
            j += 1            
plt.tight_layout()
plt.show()
```
