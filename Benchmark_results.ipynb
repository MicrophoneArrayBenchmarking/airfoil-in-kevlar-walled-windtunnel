```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
import csv

def read_csv_matrix(csv_file):
    matrix_data = []
    with open(csv_file, 'r') as f:
        reader = csv.reader(f,delimiter=';')
        for row in reader:
            row = [float(value) for value in row]
            matrix_data.append(row)
    return np.array(matrix_data)
```

## Task 1

```python
data = os.listdir('data')
task1_data = list(filter(lambda k: 'task1' in k, data))
task1_data
```

```python
fig, axs = plt.subplots(2, 2)

im1 = axs[0, 0].imshow(read_csv_matrix('data/'+task1_data[0]), cmap='viridis')
fig.colorbar(im1, ax=axs[0, 0])
axs[0, 0].set_title('CSM real 500Hz')

im2 = axs[0, 1].imshow(read_csv_matrix('data/'+task1_data[3]), cmap='viridis')
fig.colorbar(im2, ax=axs[0, 1])
axs[0, 1].set_title('CSM real 1000Hz')

im3 = axs[1, 0].imshow(read_csv_matrix('data/'+task1_data[2]), cmap='viridis')
fig.colorbar(im3, ax=axs[1, 0])
axs[1, 0].set_title('CSM imag 500Hz')

im4 = axs[1, 1].imshow(read_csv_matrix('data/'+task1_data[1]), cmap='viridis')
fig.colorbar(im4, ax=axs[1, 1])
axs[1, 1].set_title('CSM imag 1000Hz')

plt.tight_layout()
plt.show()
```
